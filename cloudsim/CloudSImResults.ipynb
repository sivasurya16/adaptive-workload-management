```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Cloudlets
cloudlets = np.array([50, 100, 150, 200, 300, 400, 500, 600, 700, 800, 900, 1000])

# Small VMs Wait Time
wait_time_small = np.array([57.14, 285.71, 700.00, 1285.71, 3000.00, 5428.57, 8571.43, 12428.57, 17000.00, 22285.71, 28285.71, 35000.00])

# Medium VMs Wait Time
wait_time_medium = np.array([20, 100, 245, 450, 1050, 1900, 3000, 4350, 5950, 7800, 9900, 12250])

# Large VMs Wait Time
wait_time_large = np.array([0.00, 5.00, 25.00, 60.00, 180.00, 360.00, 605.00, 910.00, 1280.00, 1710.00, 2205.00, 2760.00])

# Create figure and axis
plt.figure(figsize=(6, 4))

# Plot Small VMs data
plt.plot(cloudlets, wait_time_small, 'g-s', markersize=6, markerfacecolor='white', label='Small VMs')

# Plot Medium VMs data
plt.plot(cloudlets, wait_time_medium, 'y-^', markersize=6, markerfacecolor='white', label='Medium VMs')

# Plot Large VMs data
plt.plot(cloudlets, wait_time_large, 'r-o', markersize=6, markerfacecolor='white', label='Large VMs')

# Define thresholds
threshold1 = 320
threshold2 = 640

# Calculate y-values at the thresholds
small_vm_at_threshold1 = np.interp(threshold1, cloudlets, wait_time_small)
medium_vm_at_threshold1 = np.interp(threshold1, cloudlets, wait_time_medium)
medium_vm_at_threshold2 = np.interp(threshold2, cloudlets, wait_time_medium)
large_vm_at_threshold2 = np.interp(threshold2, cloudlets, wait_time_large)

# Create x and y arrays for the proposed model with sharp transitions
proposed_x = []
proposed_y = []

# Before threshold1 - follow small VM line
x_before_t1 = cloudlets[cloudlets < threshold1]
y_before_t1 = wait_time_small[cloudlets < threshold1]
proposed_x.extend(x_before_t1)
proposed_y.extend(y_before_t1)

# Add threshold1 point (on small VM line)
proposed_x.append(threshold1)
proposed_y.append(small_vm_at_threshold1)

# Add threshold1 point (on medium VM line) - creates vertical line
proposed_x.append(threshold1)
proposed_y.append(medium_vm_at_threshold1)

# Between threshold1 and threshold2 - follow medium VM line
x_between = cloudlets[(cloudlets > threshold1) & (cloudlets < threshold2)]
y_between = wait_time_medium[(cloudlets > threshold1) & (cloudlets < threshold2)]
proposed_x.extend(x_between)
proposed_y.extend(y_between)

# Add threshold2 point (on medium VM line)
proposed_x.append(threshold2)
proposed_y.append(medium_vm_at_threshold2)

# Add threshold2 point (on large VM line) - creates vertical line
proposed_x.append(threshold2)
proposed_y.append(large_vm_at_threshold2)

# After threshold2 - follow large VM line
x_after_t2 = cloudlets[cloudlets > threshold2]
y_after_t2 = wait_time_large[cloudlets > threshold2]
proposed_x.extend(x_after_t2)
proposed_y.extend(y_after_t2)

# Convert to numpy arrays
proposed_x = np.array(proposed_x)
proposed_y = np.array(proposed_y)

# Plot proposed model line with sharp transitions
plt.plot(proposed_x, proposed_y, 'b-o', markersize=6, markerfacecolor='white',linewidth=2, label='Proposed Model')
plt.plot(proposed_x, proposed_y,markersize=6)

# Formatting the axes
plt.xlabel("Cloudlets/s", fontsize=12)
plt.ylabel("Wait time (ms)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title("Wait time analysis for variable workloads")

# Scale format for y-axis (scientific notation)
plt.ticklabel_format(style='sci', axis='y', scilimits=(4,4))
plt.grid(True, linestyle="--", alpha=0.6)

# # Add annotations for thresholds
# plt.text(threshold1+5, plt.ylim()[1]*0.9, f'Threshold 1\n({threshold1})', verticalalignment='top')
# plt.text(threshold2+5, plt.ylim()[1]*0.9, f'Threshold 2\n({threshold2})', verticalalignment='top')

# Legend
plt.legend(loc="upper left", fontsize=10)

# Show the plot
plt.show()
```
